## Griegas para opción PUT Europea
**Formula de Black \& Scholes para el precio justo de una opción tipo PUT tipo Europea.**

\begin{equation}
\hat{E}[P_{t}^{LP}(t,S_{t})]=k e^{-r(T-t)}N(-d2)- S_{t}N(-d1)
\end{equation}

Consideramos que el portafolio tiene una opción put ${Q_{1}= 1}$ donde:

$$f_{t}=e^{-r(T-t)}\hat{E}[P_{t}^{LP}(t,S_{t})]$$

Si $X=ln(S_t)$:

$$ln(S_t)\sim N[(ln(S_{t}) + (\mu -\frac{\sigma^2}{2})(T-t)) , \sigma^{2} (T-t)]$$

Podemos definir el portafolio de la siguiente manera:

$$V_{t}=e^{-r(T-t)}[\int_{-\infty}^{ln(k)} (k-S_{t})p(x) dx + \int_{ln(k)}^{\infty} (0)p(x) dx ]$$

Remplazando $p(x)$:
$$V_{t}=e^{-r(T-t)}[\int_{-\infty}^{ln(k)} (k-S_{t})[ \frac{1}{\sqrt{2\pi\sigma^{2}(T-t)}}e^{\frac{-1}{2}(\frac{x-[ln(S_{t})+ (r-\frac{\sigma^2}{2})(T-t)]}  {\sigma\sqrt{(T-t)}})^2}] dx$$

### Delta $\frac{\partial V_t}{\partial S_t}= \triangle$

Previamente probamos que el precio justo de una opción Europea put está dada por la ecuación 1.

$$\hat{E}[P_{t}^{LP}(t,S_{t})]=k e^{-r(T-t)}N(-d2)- S_{t}N(-d1)$$

donde:

$$-d1=\frac{-ln(\frac{S_t}{k}) +(r+\frac{\sigma^2}{2})(T-t) }{\sigma\sqrt{(T-t)}}$$

$$-d2=\frac{-ln(\frac{S_t}{k}) +(r-\frac{\sigma^2}{2})(T-t) }{\sigma\sqrt{(T-t)}}$$

$$-d2=-d1-\sigma\sqrt{(T-t)}$$

Podemos expresar el portafolio en términos de -d1:

\begin{equation}
V_{t}=ke^{-r(T-t)} N(-d1-\sigma\sqrt{(T-t)})S_{t}N(-d1)
\end{equation}

Dividimos el portafolio respecto de -d1:

$$\frac{\partial{V_{t}}}{\partial{-d1}}=ke^{-r(T-t)}N'(-d1-\sigma\sqrt{(T-t)})S_{t}N'(-d1)$$

$$0=ke^{-r(T-t)}N'(-d1-\sigma\sqrt{(T-t)})S_{t}N'(-d1)$$

Sabemos que :

$$N'(-d1-\sigma\sqrt{(T-t)})=-d2$$

Por lo tanto,  **propiedad 1**:

$$S_{t}N'(-d1)=ke^{-r(T-t)}N'(-d2)$$

Ahora, si derivamos -d1 y -d2 con respecto de $S_{t}$ tenemos:


$$-d1=\frac{-ln(\frac{S_t}{k}) +(r+\frac{\sigma^2}{2})(T-t) }{\sigma\sqrt{(T-t)}}$$

Obsevamos que $(r+\frac{\sigma^2}{2})(T-t) $no tiene $ S_{t}$


$$\frac{\partial{-d1}}{\partial{S_t}}=\frac{-1}{S_{t}\sigma\sqrt{(T-t)}}$$


$$-d2=\frac{-ln(\frac{S_t}{k}) +(r-\frac{\sigma^2}{2})(T-t) }{\sigma\sqrt{(T-t)}}$$

$$\frac{\partial{-d2}}{\partial{S_{t}}}=\frac{-1}{S_{t}\sigma\sqrt{(T-t)}}$$

Obtenemos **propiedad 2**:
$$\frac{\partial -d1}{\partial S_t}=\frac{\partial-d2}{\partial S_t}$$

Retomando los conceptos anteriores, el valor del portafolio esta dado por la ecuación:

\begin{equation}
V_{t}=ke^{-r(T-t)} N(-d2)S_{t}N(-d1)
\end{equation}

Derivamos con respecto de $S_t$ y usando la regla de la cadena:

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}=(ke^{-r(T-t))} (\frac{\partial{N(-d2)}}{\partial{-d2}}) (\frac{\partial{-d2}}{\partial{S_t}}) - [\frac{\partial{S_t}}{\partial{S_t}} N(-d1) + (S_t)(\frac{\partial N(-d1)}{\partial -d1})(\frac{\partial-d1}{\partial S_t})]$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}=(ke^{-r(T-t))}N'(-d2)(\frac{\partial{-d2}}{\partial{S_t}}) - [(1)N(-d1)+ S_{t}N'(-d1)()] (\frac{\partial-d1}{\partial S_t})$$

Por la propiedad 2:
$$\frac{\partial -d1}{\partial S_t}=\frac{\partial -d2}{\partial S_t}$$

Obtenemos lo siguiente:

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= (\frac{\partial-d1}{\partial S_t})[ke^{-r(T-t))} N'(-d2) - S_{t}N'(-d1)] - N(-d1)$$

Por la propiedad 1:

$$S_{t}N'(-d1)=ke^{-r(T-t)}N'(-d2)$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= (\frac{\partial-d1}{\partial S_t})[ke^{-r(T-t))} N'(-d2) - ke^{-r(T-t))} - N'(-d2)]-N-(d1)$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= (\frac{\partial-d1}{\partial S_t})[0]-N-(d1)$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= -N-(d1) - \triangle$$


### Gamma $\frac{\partial {^{2}} V_t}{\partial S^{2}_t}= \Gamma$


Veamos que pasa con la segunda derivada para -d1 y -d2 ...

Sabemos que : $$\frac{\partial -d1}{\partial S_t} = -\frac{1}{S_{t} \sigma \sqrt{(T-t)}} = \frac{\partial -d2}{\partial S_t} $$

Por lo que:

$$\frac{\partial^{2}-d1}{\partial S^{2}_t} = \frac{1}{S^{2}_{t} \sigma \sqrt{(T-t)}} $$ $$\frac{\partial^{2}-d2}{\partial S^{2}_t} = \frac{1}{S^{2}_{t} \sigma \sqrt{(T-t)}} $$

**Propiedad 3**:

$$\frac{\partial^{2}-d1}{\partial S^{2}_t} = \frac{\partial^{2}-d2}{\partial S^{2}_t} $$

Retomando la ecuación 3 que fue el resultado previo de delta:


$$V_{t}=ke^{-r(T-t)} N(-d2)S_{t}N(-d1)
$$

$$\frac{\partial^{2}V_t}{\partial S^{2}_t} = \frac{\partial \triangle}{\partial S_t}= \Gamma$$

Por propiedad 2 $(\frac{\partial -d1}{\partial S_t}=\frac{\partial -d2}{\partial S_t})$:



$$\frac{\partial \triangle}{\partial S_t}= \frac{\partial - N(-d1)}{\partial S_t}= \frac{\partial N(-d1)}{\partial {-d1}} \frac{\partial -d1}{\partial S_t}$$

$$\frac{\partial \triangle}{\partial S_t}= - N'(-d1)(\frac{-1}{S_{t}\sigma\sqrt{(T-t)}})$$

Por lo tanto:

$$\frac{\partial^{2} \triangle}{\partial S^{2}_t}= - N'(-d1)(\frac{-1}{S_{t}\sigma\sqrt{(T-t)}}) = \Gamma$$


### Rho $\frac{\partial  V_t}{\partial r}= \rho$

Derivamos -d1 y -d2 con respecto de r:
$$\frac{\partial -d1}{\partial r}= - \frac{\sqrt{(T-t)}}{\sigma}$$

$$\frac{\partial -d2}{\partial r}= - \frac{\sqrt{(T-t)}}{\sigma}$$

Obtenemos una **cuarta propiedad (4)**:


$$\frac{\partial -d1}{\partial r}=\frac{\partial -d2}{\partial r}$$

Retomando el portafolio con la ecuación 3:

$$V_t=K e^{-r(T-t)}N(-d2)- S_{t}N(-d1)$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= (\frac{\partial ke^{-r(T-t)}}{\partial r}) N(-d2) + (ke^{-r(T-t)})(\frac{\partial N(-d2)}{\partial -d2}) (\frac{\partial -d2}{\partial r}) - [S_{t} (\frac{ \partial N'(-d1)}{\partial -d1}) (\frac{\partial -d1}{\partial r})] $$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= - (T-t)ke ^{-r(T-t)} N(-d2) + ke ^{-r(T-t)} N'(-d2) \frac{\partial -d2}{\partial r} - [S_{t} N'(-d1) \frac{\partial -d1}{\partial r}]$$

Por propiedad 4 y factor común:

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= - (T-t)ke ^{-r(T-t)} N(-d2) + \frac{\partial -d1}{\partial r}[ke ^{-r(T-t)} N'(-d2)-S_{t} N'(-d1)]$$

Por propiedad 1:
$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= - (T-t)ke ^{-r(T-t)} N(-d2) + \frac{\partial -d1}{\partial r}[ke ^{-r(T-t)} N'(-d2)- ke ^{-r(T-t)} N(-d2)  ]$$

$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= - (T-t)ke ^{-r(T-t)} N(-d2) + \frac{\partial -d1}{\partial r}[0]$$

Al finalizar tenemos que:
$$\frac{\partial{V_{t}}}{\partial{S_{t}}}= - (T-t)ke ^{-r(T-t)} N(-d2) = \rho$$


### Theta $\frac{\partial V_t}{\partial \tau}= \Theta$ 

Por sencillez, definimos que $(T-t)= \tau$

Es necesario obtener la segunda derivada de d1 y d2 con respecto de $\tau$

$$-d1=- \frac{ln(\frac{S_t}{k}) + (r + \frac{\sigma^2}{2})(\tau)}{\sigma\sqrt{\tau}}$$

$$\frac{\partial -d1}{\partial \tau}= \frac{1}{2} (\frac{ln(\frac{S_t}{k})}{\sigma \tau\sqrt{\tau}}) - \frac{1}{2} (\frac{r+ \frac{\sigma^2}{2}}{\sigma \sqrt{\tau}})$$

$$\frac{\partial -d1}{\partial \tau}= \frac{1}{2\sigma\sqrt{\tau}}[\frac{ln(\frac{S_t}{k})}{\tau} - (r + \frac{\sigma^2}{2})]$$

$$-d2= - \frac{ln(\frac{S_t}{k})- (r+ \frac{\sigma^2}{2})(\tau)}{\sigma\sqrt{\tau}}$$

$$-d2 =-d1 -\sigma\sqrt{\tau} \rightarrow \frac{\partial(-d1-\sigma\sqrt{\tau})}{\partial \tau}=\frac{\partial -d1}{\partial \tau} - \frac{1}{2 } \frac{\sigma}{\sqrt{\tau}}$$

**Propiedad 5**:


$$\frac{\partial -d1}{\partial \tau} = - [\frac{\partial -d2}{\partial \tau} + \frac{1}{2}\frac{\sigma}{\sqrt{\tau}}]$$

Retomando el valor del portafolio (ecuación 3):

$$V_t=K e^{-r(T-t)}N(-d2)- S_{t}N(-d1)$$

$$\frac{\partial V_t}{\partial\tau}= \frac{\partial(ke^{-r(\tau)})}{\partial \tau}N(-d2) + ke^{-r(\tau)} \frac{\partial N(-d2)}{\partial -d2} \frac{\partial -d2}{\partial \tau} - [S_{t} \frac{\partial N(-d1)}{\partial -d1} \frac{\partial -d1}{\partial \tau}]$$


$$\frac{\partial V_t}{\partial\tau}=-rke^{-r(\tau)} N(-d2) + ke^{-r(\tau)}N(-d2) \frac{\partial -d2}{\partial \tau} -[S_{t} \frac{\partial N(-d1)}{\partial -d1} \frac{\partial -d1}{\partial \tau}]$$

Por propiedad 5 y factor común:

$$\frac{\partial V_t}{\partial\tau}=-rke^{-r(\tau)} N(-d2) - \frac{1}{2} \frac{\sigma}{\sqrt{\tau}}[-S_{t}N'(-d1)] - \frac{\partial -d2}{\partial \tau}[-S_{t}N'(d1) + ke^{-r(\tau)}N'(-d2)]$$

Por propiedad 1:


$$\frac{\partial V_t}{\partial\tau}=-rke^{-r(\tau)} N(-d2) + \frac{\sigma}{2\sqrt{\tau}}S_{t}N'(d1) -\frac{\partial -d2}{\partial \tau} [-ke^{-r(\tau)}N'(-d2) + ke^{-r(\tau)}N'(-d2)]$$

$$\frac{\partial V_t}{\partial\tau}= -rke^{-r(\tau)}N(-d2) + \frac{\sigma}{2\sqrt{\tau}}S_{t}N'(-d1) - \frac{\partial -d2}{\partial \tau}[0] $$

Por naturaleza de $\tau$:

$$\frac{\partial V_t}{\partial\tau} \rightarrow \frac{-\partial V_t}{\partial\tau} \leq 0 = \Theta$$

Por lo tanto:

$$\Theta = -\frac{\partial V_t}{\partial\tau} =rke^{-r(\tau)} N(-d2) - \frac{\sigma}{2\sqrt{\tau}} S_{t}N'(-d1) $$

### Vega $\frac{\partial V_t}{\partial \sigma}= \nu$ 


Obtenemos la derivada de -d1 y -d2 con respecto de $\sigma$:

$$\frac{\partial -d1}{\partial \sigma} = \frac{ln(\frac{S_t}{k})}{\sigma^{2}\sqrt{(T-t)}} + \frac{r (T-t)}{\sigma^{2}\sqrt{(T-t)}} - \frac{(T-t)}{\sqrt{(T-t)}}$$

$$-d2= -d1 - \sigma\sqrt{(T-t)} \rightarrow \frac{\partial -d2}{\partial\sigma}= \frac{\partial -d1}{\partial \sigma}-\sqrt{(T-t)}$$

**Propiedad 6**:

$$\frac{\partial -d1}{\partial \sigma}=-[\frac{\partial -d2}{\partial\sigma} + \sqrt{(T-t)}]$$

Retomando el valor del portafolio (ecuación 3):

$$V_t=K e^{-r(T-t)}N(-d2)- S_{t}N(-d1)$$

$$\frac{\partial V_t}{\partial\sigma}= ke^{-r(T-t)} \frac{\partial N(-d2)}{\partial-d2} \frac{\partial -d2}{\partial\sigma} - S_{t} \frac{\partial N(-d1)}{\partial-d1} \frac{\partial -d1}{\partial\sigma}$$


$$\frac{\partial V_t}{\partial\sigma}= ke^{-r(T-t)} N'(-d2) \frac{\partial -d2}{\partial\sigma} - S_{t} N'(-d1) \frac{\partial -d1}{\partial\sigma}$$

Por propiedad 6 y factor común:

$$\frac{\partial V_t}{\partial\sigma}= - \frac{\partial -d2}{\partial\sigma}[ke^{-r(T-t)} N'(-d2)  - S_{t} N'(-d1)] - [S_{t} N'(-d1) (-\sqrt{(T-t})]$$

Por propiedad 1:


$$\frac{\partial V_t}{\partial\sigma}= - \frac{\partial -d2}{\partial\sigma}[ke^{-r(T-t)} N'(-d2)  -  ke^{-r(T-t)} N'(-d2)] + S_{t} N'(-d1) (\sqrt{(T-t})$$


$$\frac{\partial V_t}{\partial\sigma}= - \frac{\partial -d2}{\partial\sigma}[0] + S_{t} N'(-d1) (\sqrt{(T-t})$$



$$\frac{\partial V_t}{\partial\sigma}=  [S_{t} N'(-d1) (\sqrt{(T-t})] = \nu$$
